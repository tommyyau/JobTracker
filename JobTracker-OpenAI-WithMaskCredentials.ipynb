```python
# Importing necessary libraries
import re
import imaplib
import email
from datetime import datetime
from typing import List, Set, Tuple, Optional, Dict
from google.oauth2.service_account import Credentials
import openai
import googleapiclient.discovery
from email.header import decode_header
import time

# Configuration Constants
class Config:
    # Follow instructions to create a Google Service Account JSON Key
    SCOPES = ['https://www.googleapis.com/auth/spreadsheets']
    SERVICE_ACCOUNT_FILE = '/Users/<username>/google-service-acc-9123b9fb66d8.json'    
    SPREADSHEET_ID = '1aFindIDInURLAfterSpreadsheetsSlashDSlashID'

    # Follow instructions to create an OpenAPI Key, Caution if you create a new org - Key gets invalidated
    OPENAI_API_KEY = "sk-proj-AndARealllllyLoooooooooooooooooooogKey"

    # Follow instructions to create a Yahoo App Password, You can't use your normal login password
    EMAIL_SERVER = 'imap.mail.yahoo.com'
    EMAIL_USER = 'youremail@yahoo.com'
    EMAIL_PASSWORD = 'yahooapppassword'

    # Set the number of emails to check, it's mostly reverse chronological order but EmailID can change when you move them between folders
    MAX_EMAILS = 50
    # If you are processing a lot of emails - say 5000, then this will batch ity by disconnecting and reconnecting the service
    MAX_BATCH_SIZE = 200

    # Job Application Detection Phrases
    SUBJECT_KEY_PHRASES = [
        "your application",
        "thank you for applying",
        "we have received your application",
        "your application has been received"
    ]

    BODY_KEY_PHRASES = [
        "thank you for your application",
        "thank you for your interest",
        "thank you for applying",
        "your application has been received",
        "reviewed your application"
    ]

# Email Service
class EmailService:
    def __init__(self):
        self.mail = None

    def connect(self):
        """Establish connection to email server."""
        self.mail = imaplib.IMAP4_SSL(Config.EMAIL_SERVER)
        self.mail.login(Config.EMAIL_USER, Config.EMAIL_PASSWORD)
        self.mail.select('inbox')
        return self.mail

    def get_latest_email_ids(self, num_emails=Config.MAX_EMAILS):
        """Retrieve latest email IDs."""
        result, data = self.mail.search(None, 'ALL')
        if result != 'OK':
            print("Failed to retrieve emails")
            return []
        email_ids = data[0].split()
        return email_ids[-num_emails:]

    def extract_email_content(self, msg_data):
        """Extract email content from message data."""
        msg = email.message_from_bytes(msg_data)
        subject = self.decode_subject(msg['subject'])
        sender = msg['from']
        sender_email = self.extract_email(sender)
        date_received = msg['date']

        body = ""
        if msg.is_multipart():
            for part in msg.walk():
                if part.get_content_type() == 'text/plain' and not body:
                    body = part.get_payload(decode=True).decode('utf-8', errors='ignore')
                elif part.get_content_type() == 'text/html' and not body:
                    body = part.get_payload(decode=True).decode('utf-8', errors='ignore')
        else:
            body = msg.get_payload(decode=True).decode('utf-8', errors='ignore')

        body = re.sub(r'[^\x00-\x7F]+', '', body)
        return subject, sender_email, date_received, body

    def decode_subject(self, subject):
        """Decode email subject."""
        decoded_parts = decode_header(subject)
        decoded_subject = ''
        for part, encoding in decoded_parts:
            if isinstance(part, bytes):
                decoded_subject += part.decode(encoding or 'utf-8')
            else:
                decoded_subject += part
        return self.clean_text(decoded_subject)

    def extract_email(self, sender):
        """Extract email address from sender string."""
        match = re.search(r'<([^>]+)>', sender)
        return match.group(1) if match else sender

    def clean_text(self, text):
        """Remove non-ASCII characters from text."""
        return re.sub(r'[^\x00-\x7F]+', '', text)

# Google Sheets Service
class GoogleSheetsService:
    def __init__(self):
        self.service = self._initialize_service()

    def _initialize_service(self):
        """Initialize Google Sheets service."""
        creds = Credentials.from_service_account_file(
            Config.SERVICE_ACCOUNT_FILE,
            scopes=Config.SCOPES
        )
        return googleapiclient.discovery.build('sheets', 'v4', credentials=creds)

    def get_existing_entries(self):
        """Retrieve existing entries from Google Sheets."""
        result = self.service.spreadsheets().values().get(
            spreadsheetId=Config.SPREADSHEET_ID,
            range='A1:E'
        ).execute()
        values = result.get('values', [])
        return {(row[4], row[3]) for row in values[1:]}

    def update_sheet(self, row_data):
        """Add new row to Google Sheets."""
        self.service.spreadsheets().values().append(
            spreadsheetId=Config.SPREADSHEET_ID,
            range='A1',
            valueInputOption='RAW',
            body={'values': [row_data]}
        ).execute()

# OpenAI Service
class OpenAIService:
    def __init__(self):
        openai.api_key = Config.OPENAI_API_KEY

    def analyze_text(self, text, is_job_related=True):
        """Analyze text using OpenAI API."""
        prompt = self._create_prompt(text, is_job_related)
        try:
            response = openai.chat.completions.create(
                model="gpt-4o-mini",
                messages=[{"role": "user", "content": prompt}],
                max_tokens=100 if is_job_related else 10,
                n=1,
                stop=None,
                temperature=0.1 if is_job_related else 0.3,
            )
            return response.choices[0].message.content.strip()
        except Exception as e:
            print(f"OpenAI API error: {e}")
            return ""

    def _create_prompt(self, text, is_job_related):
        """Create appropriate prompt based on analysis type."""
        if is_job_related:
            return (
                "You are a job application email analyzer. Follow these steps:\n\n"
                "1. Read the email text carefully\n"
                "2. Extract exactly these fields in this format:\n"
                "Job Title: [Extract the specific job title. If unclear, use 'Unknown Position']\n"
                "Company Name: [Extract the company name. If unclear, use 'Unknown Company']\n"
                "Current Status: [Choose ONLY ONE: 'Applied', 'Rejected', 'Interview Request', or 'Viewed']\n\n"
                "Rules:\n"
                "- Be precise and concise\n"
                "- Only output the three fields above\n"
                "- Do not include any other text or explanations\n"
                "- If you can't determine a field, use 'Unknown' as the value\n\n"
                f"Email text:\n{text}"
            )
        else:
            return (
                "Analyze this email text and answer with ONLY 'yes' or 'no':\n"
                "Rules:\n"
                "- Answer 'yes' if the email acknowledges receiving a job application\n"
                "- Answer 'yes' if it's an update about an existing single application\n"
                "- Answer 'no' if it's promotional or unrelated to job applications\n"
                "- Answer 'no' if it's a generic update about multiple applications"
                "- Only output 'yes' or 'no', nothing else\n\n"
                f"Email text:\n{text}"
            )

# Text Processing Utilities
class TextProcessor:
    def contains_key_phrases(self, text, phrases):
        """Check if text contains any key phrases."""
        return any(phrase.lower() in text.lower() for phrase in phrases)

# Main Job Application Processing Class
class JobApplicationProcessor:
    def __init__(self):
        self.email_service = EmailService()
        self.sheets_service = GoogleSheetsService()
        self.openai_service = OpenAIService()
        self.text_processor = TextProcessor()

    def process_emails(self, batch_size=Config.MAX_BATCH_SIZE):
        """Main method to process emails in batches."""
        mail = self.email_service.connect()
        latest_emails = self.email_service.get_latest_email_ids()
        existing_entries = self.sheets_service.get_existing_entries()

        print(f"Processing {len(latest_emails)} emails")

        for start in range(0, len(latest_emails), batch_size):
            self._process_batch(mail, latest_emails[start:start + batch_size], existing_entries)
            mail.logout()
            print(f"Completed batch processing: {start+batch_size}/{len(latest_emails)}")
            mail = self.email_service.connect()

    def _process_batch(self, mail, email_batch, existing_entries):
        """Process a batch of emails."""
        total_emails = len(email_batch)
        for index, email_id in enumerate(email_batch, 1):
            try:
                progress_info = f"[{index} of {total_emails}]"
                self._process_single_email(mail, email_id, existing_entries, progress_info)
                time.sleep(0.1)  # Rate limiting
            except Exception as e:
                print(f"[{index} of {total_emails}] Error: {e}")

    def _process_single_email(self, mail, email_id, existing_entries, progress_info):
        """Process a single email."""
        result, msg_data = mail.fetch(email_id, '(RFC822)')
        if result != 'OK':
            return

        subject, sender_email, date_received, body = self.email_service.extract_email_content(msg_data[0][1])
        combined_text = f"{subject}\n{body}"

        if self._is_job_related_email(subject, body):
            self._process_job_related_email(combined_text, subject, sender_email,
                                        date_received, existing_entries, progress_info)
        else:
            print(f"{progress_info} Not job-related: {subject[:50]}...")


    def _is_job_related_email(self, subject, body):
        """Determine if email is job-related."""
        if self.text_processor.contains_key_phrases(subject, Config.SUBJECT_KEY_PHRASES):
            return True
        if self.text_processor.contains_key_phrases(body, Config.BODY_KEY_PHRASES):
            return True
        return 'yes' in self.openai_service.analyze_text(f"{subject}\n{body}", False).lower()

    def _process_job_related_email(self, combined_text, subject, sender_email, date_received, existing_entries, progress_info):
        """Process job-related email and update spreadsheet."""
        extracted_info = self.openai_service.analyze_text(combined_text, True)

        job_info = self._extract_job_info(extracted_info, subject, sender_email)
        if (sender_email, date_received) not in existing_entries:
            self.sheets_service.update_sheet([
                job_info['title'].replace("** ", ""),
                job_info['company'].replace("** ", ""),
                job_info['status'].replace("** ", ""),
                date_received,
                sender_email
            ])
            print(f"{progress_info} Added new entry: {job_info['company']}")
        else:
            print(f"{progress_info} Duplicate entry: {job_info['company']}")

    def _extract_job_info(self, extracted_info, subject, sender_email):
        """Extract job information from OpenAI response."""
        job_title_match = re.search(r'Job Title:\s*(.+)', extracted_info)
        company_name_match = re.search(r'Company Name:\s*(.+)', extracted_info)
        current_status_match = re.search(r'Current Status:\s*(.+)', extracted_info)

        job_info = {
            'title': job_title_match.group(1).strip() if job_title_match else subject,
            'company': company_name_match.group(1).strip() if company_name_match else sender_email,
            'status': 'Not Known'
        }

        if "your application was viewed by" in subject.lower():
            job_info['status'] = "Viewed"
            job_info['company'] = subject.lower().split("your application was viewed by")[1].strip()
        elif current_status_match:
            job_info['status'] = current_status_match.group(1).strip()

        if "N/A" in job_info['company'] or "Not specified" in job_info['company'] or "Not explicitly" in job_info['company']:
            job_info['company'] = sender_email

        return job_info

# Create and run the processor

processor = JobApplicationProcessor()
processor.process_emails()
```
